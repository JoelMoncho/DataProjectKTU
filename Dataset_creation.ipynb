## Dataset Creation

#### Read CSVs

```python
import pandas as pd

df = pd.read_csv("Data/nrg_bal_c_linear_2_0.csv", encoding='utf-8')  

```

```python
df.head()
```

#### General Dataframe

```python
df_general = df[['geo', 'TIME_PERIOD', 'nrg_bal', 'OBS_VALUE']]

variables_interes = [
    'GEP', 'GEP_MAPE', 'GEP_MAPCHP',
    'GHP', 'GHP_MAPCHP',
    'FC_E', 'FC_IND_E', 'FC_TRA_E', 'FC_OTH_E',
    'AFC', 'DL'
]
df_general = df_general[df_general['nrg_bal'].isin(variables_interes)]


df_general = df_general.pivot_table(
    index=['geo', 'TIME_PERIOD'],
    columns='nrg_bal',
    values='OBS_VALUE',
    aggfunc='sum'
).reset_index()

df_general.to_csv("df_general.csv", index=False)
df_general
```

##### Resume df_general

```python

column_split = [col.split('_', 1) if '_' in col else [col, ''] for col in df_general.columns]

df_columns = pd.DataFrame(column_split, columns=['Variable', 'Combustible'])

print("Número de columnas por variable:")
print(df_columns.groupby('Variable').size())

print("\nLista de combustibles presentes:")
print(df_columns['Combustible'].unique())

print("\nTodas las columnas del DataFrame:")
print(df_general.columns.tolist())
```

#### Specific Dataframe

```python
print(df['siec'].unique())
```

```python
df_siec = df[['geo', 'TIME_PERIOD', 'nrg_bal', 'siec', 'OBS_VALUE']]

variables_interes = [
    'GEP', 'GEP_MAPE', 'GEP_MAPCHP',
    'GHP', 'GHP_MAPCHP',
    'FC_E', 'FC_IND_E', 'FC_TRA_E', 'FC_OTH_E',
    'AFC', 'DL'
]
df_siec = df_siec[df_siec['nrg_bal'].isin(variables_interes)]

siec_map = {
    'BIOE': 'Bioenergía',
    'C0000X0350-0370': 'Petróleo',
    'C0110': 'Petróleo',
    'C0121': 'Petróleo',
    'C0129': 'Petróleo',
    'C0210': 'Gas',
    'C0220': 'Gas',
    'C0311': 'Carbón',
    'C0312': 'Carbón',
    'C0320': 'Carbón',
    'C0330': 'Carbón',
    'C0340': 'Carbón',
    'C0350': 'Petróleo',
    'RA000': 'Nuclear',
    'RA100': 'Nuclear',
    'RA200': 'Nuclear',
    'RA300': 'Nuclear',
    'RA410': 'Nuclear',
    'RA420': 'Nuclear',
    'RA500': 'Nuclear',
    'RA600': 'Nuclear',
    'S2000': 'Renovables_total',
    'W6100': 'Hidráulica',
    'W6210': 'Eólica',
    'W6220': 'Solar'
}
df_siec = df_siec[df_siec['siec'].isin(siec_map.keys())]
df_siec['siec'] = df_siec['siec'].map(siec_map)

df_siec = df_siec.pivot_table(
    index=['geo', 'TIME_PERIOD'],
    columns=['nrg_bal', 'siec'],
    values='OBS_VALUE',
    aggfunc='sum'
)

df_siec.columns = [f'{var}_{siec}' for var, siec in df_siec.columns]
df_siec.reset_index(inplace=True)

df_siec.to_csv("df_siec.csv", index=False)
df_siec
```

##### Creating totals

```python

afc_cols = [col for col in df_siec.columns if col.startswith('AFC_')]
df_siec['AFC_total'] = df_siec[afc_cols].sum(axis=1)

dl_cols = [col for col in df_siec.columns if col.startswith('DL_')]
df_siec['DL_total'] = df_siec[dl_cols].sum(axis=1)

fc_cols = [col for col in df_siec.columns if col.startswith('FC_E_')]
df_siec['FC_E_total'] = df_siec[fc_cols].sum(axis=1)

fc_ind_cols = [col for col in df_siec.columns if col.startswith('FC_IND_E_')]
df_siec['FC_IND_total'] = df_siec[fc_ind_cols].sum(axis=1)

fc_tra_cols = [col for col in df_siec.columns if col.startswith('FC_TRA_E_')]
df_siec['FC_TRA_total'] = df_siec[fc_tra_cols].sum(axis=1)

fc_oth_cols = [col for col in df_siec.columns if col.startswith('FC_OTH_E_')]
df_siec['FC_OTH_total'] = df_siec[fc_oth_cols].sum(axis=1)

df_siec.to_csv("df_siec.csv", index=False)
df_siec
```

##### Resume df_siec

```python


variables_base = ['AFC', 'DL', 'FC_E', 'FC_IND_E', 'FC_OTH_E', 'FC_TRA_E',
                  'GEP', 'GEP_MAPCHP', 'GEP_MAPE', 'GHP', 'GHP_MAPCHP', 
                  'TIME_PERIOD', 'geo']

column_split = []
for col in df_siec.columns:
    var_match = next((v for v in variables_base if col.startswith(v)), col)
    
    if '_' in col and var_match != col:
        combustible = col.split('_')[-1]
    else:
        combustible = ''

    if var_match.startswith('FC'):
        var_group = 'FC'
    elif var_match.startswith('AFC'):
        var_group = 'AFC'
    elif var_match.startswith('DL'):
        var_group = 'DL'
    elif var_match.startswith('GEP'):
        var_group = 'GEP'
    elif var_match.startswith('GHP'):
        var_group = 'GHP'
    elif var_match == 'TIME_PERIOD':
        var_group = 'TIME'
    elif var_match == 'geo':
        var_group = 'geo'
    else:
        var_group = var_match
    
    column_split.append([var_group, combustible])


df_columns = pd.DataFrame(column_split, columns=['Variable', 'Combustible'])

print("Número de columnas por variable:")
print(df_columns.groupby('Variable').size())

print("\nLista de combustibles presentes:")
print(df_columns['Combustible'].unique())

print("\nTodas las columnas del DataFrame:")
print(df_siec.columns.tolist())
```

#### Adding GDP

```python
df_gdp = pd.read_csv("Data/nama_10_gdp__custom_17992921_linear_2_0.csv", encoding='utf-8')  
df_gdp
print(df_gdp.columns.tolist())
```

```python
df_gdp_sel = df_gdp[df_gdp['unit'].isin(['CLV20_MEUR', 'CLV20_MNAC', 'CLV_PCH_PRE'])]

df_gdp_sel = df_gdp_sel[['geo', 'TIME_PERIOD', 'unit', 'OBS_VALUE']]

df_gdp_pivot = df_gdp_sel.pivot_table(
    index=['geo', 'TIME_PERIOD'],
    columns='unit',
    values='OBS_VALUE'
).reset_index()

df_gdp_pivot = df_gdp_pivot.rename(columns={
    'CLV20_MEUR': 'PIB_real_MEUR',
    'CLV20_MNAC': 'PIB_real_NAC',
    'CLV_PCH_PRE': 'PIB_growth_pct'
})

df_general = pd.merge(df_general, df_gdp_pivot, on=['geo', 'TIME_PERIOD'], how='left')
df_siec = pd.merge(df_siec, df_gdp_pivot, on=['geo', 'TIME_PERIOD'], how='left')

df_general.to_csv("df_general.csv", index=False)
df_siec.to_csv("df_siec.csv", index=False)

df_general
```

#### Addinng Emissions

```python

df_emi = pd.read_csv("Data/env_air_gge__custom_17993346_linear_2_0.csv", encoding='utf-8')  
df_emi
```

```python
df_emi = df_emi[['geo', 'TIME_PERIOD', 'airpol', 'OBS_VALUE']]

df_emi_pivot = df_emi.pivot_table(
    index=['geo', 'TIME_PERIOD'],
    columns='airpol',
    values='OBS_VALUE',
    aggfunc='sum' 
)


df_emi_pivot.columns = [f'EMI_{col}' for col in df_emi_pivot.columns]
df_emi_pivot.reset_index(inplace=True)

df_general = pd.merge(df_general, df_emi_pivot, on=['geo', 'TIME_PERIOD'], how='left')

df_siec = pd.merge(df_siec, df_emi_pivot, on=['geo', 'TIME_PERIOD'], how='left')

df_general.to_csv("df_general.csv", index=False)
df_siec.to_csv("df_siec.csv", index=False)


df_general
```

#### Adding Electricity Consumption

```python

df_cons = pd.read_csv("Data/nrg_pc_204__custom_17993638_linear_2_0.csv", encoding='utf-8')  
df_cons
```

```python
df_cons = df_cons[['geo', 'TIME_PERIOD', 'OBS_VALUE']]


df_cons['YEAR'] = df_cons['TIME_PERIOD'].str[:4].astype(int)

df_cons_annual = df_cons.groupby(['geo', 'YEAR'], as_index=False)['OBS_VALUE'].mean()
df_cons_annual.rename(columns={'YEAR':'TIME_PERIOD', 'OBS_VALUE':'CONS_ELECT_ANNUAL'}, inplace=True)
df_general = pd.merge(df_general, df_cons_annual, on=['geo', 'TIME_PERIOD'], how='left')

df_siec = pd.merge(df_siec, df_cons_annual, on=['geo', 'TIME_PERIOD'], how='left')

df_general.to_csv("df_general.csv", index=False)
df_siec.to_csv("df_siec.csv", index=False)
```

#### Adding transport data

```python
df_trans = df_trans[['geo', 'TIME_PERIOD', 'vehicle', 'OBS_VALUE']]

df_trans_pivot = df_trans.pivot_table(
    index=['geo', 'TIME_PERIOD'],
    columns='vehicle',
    values='OBS_VALUE',
    aggfunc='sum'
)
df_trans_pivot.columns = [f"TRANS_{col}" for col in df_trans_pivot.columns]
df_trans_pivot.reset_index(inplace=True)

df_general = df_general.merge(df_trans_pivot, on=['geo', 'TIME_PERIOD'], how='left')

df_siec = df_siec.merge(df_trans_pivot, on=['geo', 'TIME_PERIOD'], how='left')

df_general.to_csv("df_general.csv", index=False)
df_siec.to_csv("df_siec.csv", index=False)

```

### Variables

| Column              | Description                                                               |
| ------------------- | ------------------------------------------------------------------------- |
| `geo`               | Country code (ISO 2-letter, e.g., ES for Spain, FR for France)            |
| `TIME_PERIOD`       | Year of observation                                                       |
| `AFC`               | Final energy consumption by fuel (all sectors)                            |
| `DL`                | Domestic energy losses (distribution, transformation, etc.)               |
| `FC_E`              | Final energy consumption (total)                                          |
| `FC_IND_E`          | Final energy consumption in industry sector                               |
| `FC_OTH_E`          | Final energy consumption in other sectors (residential, services, etc.)   |
| `FC_TRA_E`          | Final energy consumption in transport sector                              |
| `GEP`               | Gross electricity production (all sources)                                |
| `GEP_MAPCHP`        | Gross electricity production from combined heat & power (CHP)             |
| `GEP_MAPE`          | Gross electricity production from renewable sources                       |
| `GHP`               | Gross heat production                                                     |
| `GHP_MAPCHP`        | Gross heat production from combined heat & power (CHP)                    |
| `PIB_real_MEUR`     | Real GDP in million euros (chain-linked volumes, reference year 2020)     |
| `PIB_real_NAC`      | Real GDP in national currency (chain-linked volumes, reference year 2020) |
| `PIB_growth_pct`    | Annual GDP growth rate (%)                                                |
| `EMI_CH4`           | Methane (CH₄) emissions                                                   |
| `EMI_CH4_CO2E`      | Methane (CH₄) emissions expressed in CO₂ equivalents                      |
| `EMI_CO2`           | Carbon dioxide (CO₂) emissions                                            |
| `EMI_GHG`           | Total greenhouse gas (GHG) emissions                                      |
| `EMI_N2O`           | Nitrous oxide (N₂O) emissions                                             |
| `EFF_PEC_EED`       | Primary energy consumption efficiency (energy efficiency indicators, EED) |
| `CONS_ELECT_ANNUAL` | Annual electricity consumption (households & industry)                    |
| `TRANS_AC`          | Passenger transport by aircraft (% of total passenger-km)                 |
| `TRANS_BUS_TOT`     | Passenger transport by bus (total, % of passenger-km)                     |
| `TRANS_CAR`         | Passenger transport by car (total, % of passenger-km)                     |
| `TRANS_SEAV`        | Passenger transport by sea & inland waterways (% of passenger-km)         |
| `TRANS_TRN`         | Passenger transport by train (% of passenger-km)                          |
