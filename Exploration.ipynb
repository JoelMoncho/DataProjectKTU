```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("df_general.csv")
```

```python
print("Shape:", df.shape)
print("Columns:", df.columns.tolist())
print("Time period:", df['TIME_PERIOD'].min(), "→", df['TIME_PERIOD'].max())
print("Number of countries:", df['geo'].nunique())

print("\nMissing values per column:\n", df.isna().sum())

df.describe().T
```

```python
country_years = (
    df.groupby("geo")["TIME_PERIOD"]
    .agg(["min", "max"])
    .reset_index()
    .rename(columns={"min": "FirstYear", "max": "LastYear"})
)

country_years["NumYears"] = (
    df.groupby("geo")["TIME_PERIOD"].nunique().values
)

country_years = country_years.sort_values("FirstYear")

print(country_years)

coverage = pd.crosstab(df["geo"], df["TIME_PERIOD"])

plt.figure(figsize=(14,8))
sns.heatmap(coverage, cmap="Greens", cbar=False)
plt.title("Temporal data coverage by country")
plt.xlabel("Year")
plt.ylabel("Country")
plt.show()
```

```python

agregados = ["EU27_2020", "EU28", "EA19", "EA20", "EA", "EFTA", "EU15", "EU25", "EU27"]


df_countries = df[~df["geo"].isin(agregados)].copy()


country_years = (
    df_countries.groupby("geo")["TIME_PERIOD"]
    .agg(["min", "max", "nunique"])
    .reset_index()
    .rename(columns={"min": "FirstYear", "max": "LastYear", "nunique": "NumYears"})
)

valid_countries = country_years[country_years["NumYears"] >= 30]["geo"]
df_countries = df_countries[df_countries["geo"].isin(valid_countries)].copy()

print("Número de países finales (sin agregados y con ≥30 años):", df_countries["geo"].nunique())
print("Ejemplo de códigos de países:", df_countries["geo"].unique()[:10])
```

```python
vars_to_plot = ["PIB_real_MEUR", "EMI_GHG", "FC_E"]

for var in vars_to_plot:
    plt.figure()
    sns.histplot(np.log1p(df_countries[var].dropna()), bins=30, kde=True)
    plt.title(f"Distribución log({var})")
    plt.xlabel(f"log({var})")
    plt.show()
```

```python
num_vars = df.select_dtypes(include=[np.number])
corr = num_vars.corr()

plt.figure(figsize=(12,8))
sns.heatmap(corr, cmap="coolwarm", center=0, annot=False)
plt.title("Matriz de correlaciones")
plt.show()
```

```python
plt.figure()
sns.scatterplot(data=df, x="PIB_real_MEUR", y="EMI_GHG", hue="geo", alpha=0.6, legend=False)
plt.title("Relación PIB vs Emisiones GHG")
plt.xlabel("PIB (millones €)")
plt.ylabel("Emisiones GHG")
plt.show()
```

```python
df_eu = df.groupby("TIME_PERIOD")[["PIB_real_MEUR","EMI_GHG","FC_E"]].sum().reset_index()

fig, ax1 = plt.subplots()
sns.lineplot(data=df_eu, x="TIME_PERIOD", y="PIB_real_MEUR", ax=ax1, label="PIB")
ax2 = ax1.twinx()
sns.lineplot(data=df_eu, x="TIME_PERIOD", y="EMI_GHG", ax=ax2, color="red", label="Emisiones")
ax1.set_ylabel("PIB (millones €)")
ax2.set_ylabel("Emisiones GHG")
plt.title("Evolución PIB y Emisiones (UE total)")
plt.show()
```

```python
# PIB por país
plt.figure(figsize=(12,6))
sns.lineplot(data=df_countries, x="TIME_PERIOD", y="PIB_real_MEUR", hue="geo", alpha=0.7, lw=1)
plt.title("Evolución del PIB por país")
plt.ylabel("PIB (millones €)")
plt.show()

# Emisiones por país
plt.figure(figsize=(12,6))
sns.lineplot(data=df_countries, x="TIME_PERIOD", y="EMI_GHG", hue="geo", alpha=0.7, lw=1)
plt.title("Evolución de Emisiones GHG por país")
plt.ylabel("Emisiones GHG")
plt.show()
```
